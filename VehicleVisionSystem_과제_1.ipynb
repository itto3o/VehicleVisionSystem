```python
from google.colab.patches import cv2_imshow
import numpy as np
import cv2
import matplotlib.pyplot as plt

#영상 불러오기
grayScale_img = cv2.imread('/content/gdrive/MyDrive/keimyeong Univ/VehicleVisionSystem/grayscale_apple.jpeg', cv2.IMREAD_GRAYSCALE)
plt.imshow(grayScale_img, 'gray')
```

```python
#찬혁님이 자료형을 붙여야 티가 난대..
grayImg_mul = (grayScale_img * 1.3).astype(np.uint8)
plt.imshow(grayImg_mul, 'gray')
```

```python
grayImg_div = (grayScale_img / 1.3).astype(np.uint16)
plt.imshow(grayImg_div, 'gray')
```

```python
grayImg_add = grayScale_img + 200
plt.imshow(grayImg_add, 'gray')
```

```python
grayImg_add_float = grayScale_img + 200.
grayImg_add_saturation = np.clip(grayImg_add_float, 0, 255).astype(np.uint8)
plt.imshow(grayImg_add_saturation, 'gray')
```

```python
#where 함수로, 조건에 만족하면(픽셀 값이 255보다 큰 값이 있으면) 256을 뺀 값을 넣고, 아니면 그대로
grayImg_add_wrapping = (np.where(grayImg_add_float > 255,  grayImg_add_float - 256, grayImg_add_float)).astype(int)
plt.imshow(grayImg_add_wrapping, 'gray')
```

```python
grayImg_add_all = grayScale_img + grayImg_mul + grayImg_div
plt.imshow(grayImg_add_all, 'gray')
```

```python
grayImg_add_all_saturation = np.clip(grayImg_add_all, 0, 255).astype(np.uint8)
plt.imshow(grayImg_add_all_saturation, 'gray')
```

```python
grayImg_add_all_wrapping = np.where(grayImg_add_all > 255,  grayImg_add_all - 256, grayImg_add_all)
plt.imshow(grayImg_add_all_wrapping, 'gray')
```

```python
binaryImg1 = cv2.imread('/content/gdrive/MyDrive/keimyeong Univ/VehicleVisionSystem/binaryImage1.png')
binaryImg2 = cv2.imread('/content/gdrive/MyDrive/keimyeong Univ/VehicleVisionSystem/binaryImage2.png')

#583, 771과 585, 780으로 img2만 맞추면 될 것 같다.
print(binaryImg1.shape, binaryImg2.shape)
binaryImg2 = cv2.resize(binaryImg2, (771, 583))
print(binaryImg2.shape)

plt.figure(figsize = (20, 20))
plt.subplot(1, 2, 1) #1행 4열의 1번째 열
plt.title("img1")
plt.imshow(binaryImg1)

plt.subplot(1, 2, 2)
plt.title("img2")
plt.imshow(binaryImg2)
```

```python
bitwise_and = cv2.bitwise_and(binaryImg1, binaryImg2)
cv2_imshow(bitwise_and)
```

```python
bitwise_or = cv2.bitwise_or(binaryImg1, binaryImg2)
cv2_imshow(bitwise_or)
```
